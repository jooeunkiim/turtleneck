```
%tensorflow_version 2.3.x
import tensorflow as tf
import os
```

```
WIDTH = 320
HEIGHT = 240
```

```
def create_model(print_sum = False):
  model = tf.keras.Sequential([
      tf.keras.layers.Conv2D(input_shape=(HEIGHT,WIDTH,3), kernel_size=(3,3), filters=32, activation='relu'),
      tf.keras.layers.MaxPool2D(strides=(2,2)),
      tf.keras.layers.Conv2D(kernel_size=(3,3), filters=64, activation='relu'),
      tf.keras.layers.MaxPool2D(strides=(2,2)),
      tf.keras.layers.Conv2D(kernel_size=(3,3), filters=128, activation='relu'),
      tf.keras.layers.MaxPool2D(strides=(2,2)),
      tf.keras.layers.Flatten(),
      tf.keras.layers.Dense(units=128, activation='relu'),
      tf.keras.layers.Flatten(),
      tf.keras.layers.Dense(units=2, activation='softmax')
  ])

  model.compile(optimizer=tf.keras.optimizers.Adam(), 
                loss='sparse_categorical_crossentropy', 
                metrics=['accuracy'])
  if print_sum:
    model.summary()
  return model
```

```
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
  '/tmp/Data/CK',
  image_size=(HEIGHT,WIDTH),
  shuffle = True,  
  seed=123)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
  '/tmp/Data/DH',
  image_size=(HEIGHT,WIDTH),
  shuffle = True,
  seed=123)
```

```
train_ds.class_names
```

```
model = create_model()

model.fit(
  train_ds,
  validation_data=val_ds,
  epochs = 10
)
```
